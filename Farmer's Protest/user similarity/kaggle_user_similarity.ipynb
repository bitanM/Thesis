# User Similarity Graph (Attribute-Based)

This notebook builds a user similarity graph from `users.csv` using account attributes and bio text.

Similarity components:
- log-normalized followers count
- log-normalized following count (friendsCount)
- account age in days at protest date
- verified status (binary)
- bio TF-IDF cosine similarity

Edges are created by connecting each user to their top-k most similar users (k=10-15).


```python
# Optional: install if needed
# !pip -q install -U pandas scikit-learn networkx tqdm

import os
import json
import numpy as np
import pandas as pd
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

try:
    from tqdm import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x

# ==== PARAMETERS ====
INPUT_PATH = '/kaggle/input/farmers-protest-data/users.csv'
PROTEST_DATE = '2020-11-26'  # update if you want a different anchor date
OUTDIR = '/kaggle/working/user_similarity'

K = 12  # pick between 10-15
K_CANDIDATES = 60  # candidates from bio similarity before final re-ranking
ALPHA_BIO = 0.6  # weight for bio similarity (0-1); rest is numeric similarity

MAX_FEATURES = 20000
MIN_DF = 2
```

```python
df = pd.read_csv(INPUT_PATH)

# Basic cleaning / defaults
df['rawDescription'] = df['rawDescription'].fillna('')
df['followersCount'] = df['followersCount'].fillna(0)
df['friendsCount'] = df['friendsCount'].fillna(0)
df['verified'] = df['verified'].fillna(False).astype(bool)

# Parse created date and compute account age in days
df['created'] = pd.to_datetime(df['created'], errors='coerce', utc=True)
protest_dt = pd.to_datetime(PROTEST_DATE, utc=True)
df['account_age_days'] = (protest_dt - df['created']).dt.days
df['account_age_days'] = df['account_age_days'].fillna(0).clip(lower=0)

print('Users:', len(df))
```

```python
# Numeric features: log-normalized counts + age + verified
followers = np.log1p(df['followersCount'].to_numpy())
following = np.log1p(df['friendsCount'].to_numpy())
age_days = df['account_age_days'].to_numpy()
verified = df['verified'].astype(int).to_numpy()

num_features = np.vstack([followers, following, age_days, verified]).T
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num_features)

# L2 normalize numeric features for cosine similarity
num_norm = np.linalg.norm(num_scaled, axis=1, keepdims=True)
num_norm[num_norm == 0] = 1.0
num_unit = num_scaled / num_norm

# Bio TF-IDF
tfidf = TfidfVectorizer(
    max_features=MAX_FEATURES,
    min_df=MIN_DF,
    stop_words='english'
)
bio_tfidf = tfidf.fit_transform(df['rawDescription'])

print('TF-IDF shape:', bio_tfidf.shape)
```

```python
# KNN over bio cosine similarity to get candidate neighbors
knn = NearestNeighbors(n_neighbors=K_CANDIDATES + 1, metric='cosine', algorithm='brute')
knn.fit(bio_tfidf)
distances, indices = knn.kneighbors(bio_tfidf, return_distance=True)

# Build edges using combined similarity
edges = {}  # (min_id, max_id) -> weight
user_ids = df['userId'].astype(str).to_numpy()

for i in tqdm(range(len(df)), desc='Building edges'):
    neigh_idx = indices[i]  # includes self at position 0
    neigh_dist = distances[i]

    candidates = []
    for rank in range(1, len(neigh_idx)):
        j = neigh_idx[rank]
        bio_sim = 1.0 - neigh_dist[rank]
        # numeric cosine similarity
        num_sim = float(np.dot(num_unit[i], num_unit[j]))
        sim = ALPHA_BIO * bio_sim + (1.0 - ALPHA_BIO) * num_sim
        candidates.append((j, sim, bio_sim, num_sim))

    # pick top-K by combined similarity
    candidates.sort(key=lambda x: x[1], reverse=True)
    topk = candidates[:K]

    for j, sim, bio_sim, num_sim in topk:
        u = user_ids[i]
        v = user_ids[j]
        a, b = (u, v) if u <= v else (v, u)
        # keep the strongest edge if duplicated
        if (a, b) not in edges or sim > edges[(a, b)]['weight']:
            edges[(a, b)] = {
                'weight': float(sim),
                'bio_sim': float(bio_sim),
                'num_sim': float(num_sim),
            }

print('Edges:', len(edges))
```

```python
# Build graph and export
G = nx.Graph()

for _, row in df.iterrows():
    uid = str(row['userId'])
    G.add_node(
        uid,
        username=row.get('username', ''),
        displayname=row.get('displayname', ''),
        verified=bool(row.get('verified', False)),
        followersCount=int(row.get('followersCount', 0)),
        friendsCount=int(row.get('friendsCount', 0)),
        account_age_days=int(row.get('account_age_days', 0)),
    )

for (u, v), attrs in edges.items():
    G.add_edge(u, v, **attrs)

os.makedirs(OUTDIR, exist_ok=True)
edges_path = os.path.join(OUTDIR, 'user_similarity_edges.csv')
nodes_path = os.path.join(OUTDIR, 'user_similarity_nodes.csv')
gexf_path = os.path.join(OUTDIR, 'user_similarity_graph.gexf')

edges_df = pd.DataFrame([
    {'source': u, 'target': v, **attrs}
    for (u, v), attrs in edges.items()
])
edges_df.to_csv(edges_path, index=False)

nodes_df = pd.DataFrame({
    'userId': list(G.nodes()),
    'username': [G.nodes[n].get('username', '') for n in G.nodes()],
    'displayname': [G.nodes[n].get('displayname', '') for n in G.nodes()],
    'verified': [G.nodes[n].get('verified', False) for n in G.nodes()],
    'followersCount': [G.nodes[n].get('followersCount', 0) for n in G.nodes()],
    'friendsCount': [G.nodes[n].get('friendsCount', 0) for n in G.nodes()],
    'account_age_days': [G.nodes[n].get('account_age_days', 0) for n in G.nodes()],
})
nodes_df.to_csv(nodes_path, index=False)

nx.write_gexf(G, gexf_path)

stats = {
    'users': len(G.nodes()),
    'edges': len(G.edges()),
    'k': K,
    'alpha_bio': ALPHA_BIO,
}
with open(os.path.join(OUTDIR, 'stats.json'), 'w') as f:
    json.dump(stats, f, indent=2)

print('Done:', stats)
```
